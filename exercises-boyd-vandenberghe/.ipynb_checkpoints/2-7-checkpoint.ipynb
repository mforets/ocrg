```python
a = random_vector(RR, 2, min = -1, max = 1)
b = random_vector(RR, 2, min = -1, max = 1)
```

```python
norm2 = lambda x : sqrt(sum(xi*xi for xi in x))
```

```python
d = (norm2(b)^2-norm2(a)^2)/2
c = b - a
```

```python
# plot the points a and b
drawing = point(a, color='red', marker='x') + point(b, color='blue',marker='x')

# plot the region, with some (optional) formatting tweaks
drawing += region_plot(lambda x, y: c[0]*x + c[1]*y <= d ,(-1,1),(-1,1), 
                       incol='gold', bordercol='black', borderstyle='dashed', 
                       plot_points=500, aspect_ratio=1, alpha=0.3)

drawing.show()
```

---

```python
var('x y')
eq = c[0]*x + c[1]*y == d
sol = solve(eq, y)
h1(x) = sol[0].rhs()

plot(h1, -1, 1, fill=h1(1),fillcolor='lightgrey')
```

### Defining a polyhedron 

```python
from lib.polyFunctions_core import *

a, b = random_vector(RR, 2, min = -1, max = 1), random_vector(RR, 2, min = -1, max = 1)

norm2 = lambda x : sqrt(sum(xi*xi for xi in x))

d = vector([(norm2(b)^2-norm2(a)^2)/2])
c = matrix([b - a])

P = PolyhedronFromHSpaceRep(c, d, base_ring = RDF)

print 'Is P bounded?', P.is_compact()
print 'The inequalities defining P are ', P.inequalities()
```

```python
G = Graphics()

# plot the points, with some formatting
G = point(a, color='red', marker='x') + point(b, color='blue',marker='x')

# plot the region, note that it is input as an inline function
G += region_plot(lambda x, y: c[0]*x + c[1]*y <= d ,(-1,1),(-1,1), alpha=0.1, color='blue', borderstyle='solid')
G 
```
